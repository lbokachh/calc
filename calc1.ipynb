```python
dict_chisla = {'и': -1, 'десятых': 0.1, 'сотых': 0.01, 'тысячных': 0.001,
    'десятая': 0.1, 'сотая': 0.01, 'тысячная': 0.001,
    'ноль': 0, 'один': 1, 'одна': 1, 'два': 2, 'две': 2, 'три': 3, 'четыре': 4,
    'пять': 5, 'шесть': 6, 'семь': 7, 'восемь': 8,
    'девять': 9, 'десять': 10, 'одиннадцать': 11, 'двенадцать': 12,
    'тринадцать': 13, 'четырнадцать': 14, 'пятнадцать': 15, 'шестнадцать': 16,
    'семнадцать': 17, 'восемнадцать': 18, 'девятнадцать': 19, 'двадцать': 20,
    'тридцать': 30, 'сорок': 40, 'пятьдесят': 50, 'шестьдесят': 60,
    'семьдесят': 70, 'восемьдесят': 80, 'девяносто': 90, 'сто': 100
}

dict_obrat_chisla = {
    '0': 'ноль', '1': 'один', '2': 'два', '3': 'три',
    '4': 'четыре', '5': 'пять', '6': 'шесть', '7': 'семь',
    '8': 'восемь', '9': 'девять', '10': 'десять', '11': 'одиннадцать',
    '12': 'двенадцать', '13': 'тринадцать', '14': 'четырнадцать', '15': 'пятнадцать',
    '16': 'шестнадцать', '17': 'семнадцать', '18': 'восемнадцать', '19': 'девятнадцать',
    '20': 'двадцать', '30': 'тридцать', '40': 'сорок', '50': 'пятьдесят',
    '60': 'шестьдесят', '70': 'семьдесят', '80': 'восемьдесят', '90': 'девяносто',
    '100': 'сто', '200': 'двести', '300': 'тристо', '400': 'четыресто',
    '500': 'пятьсот', '600': 'шестьсот', '700': 'семьсот', '800': 'восемьсот',
    '900': 'девятьсот', '1000': 'тысяча', '2000': 'две тысячи', '3000': 'три тысячи',
    '4000': 'четыре тысячи', '5000': 'пять тысяч', '6000': 'шесть тысяч', '7000': 'семь тысяч',
    '8000': 'восемь тысяч', '9000': 'девять тысяч', '10000': 'десять тысяч'
}

dict_obrat_chisla_drob = {
    '0': '', '1': 'одна', '2': 'две', '3': 'три',
    '4': 'четыре', '5': 'пять', '6': 'шесть', '7': 'семь',
    '8': 'восемь', '9': 'девять', '10': 'десять', '11': 'одиннадцать',
    '12': 'двенадцать', '13': 'тринадцать', '14': 'четырнадцать', '15': 'пятнадцать',
    '16': 'шестнадцать', '17': 'семнадцать', '18': 'восемнадцать', '19': 'девятнадцать',
    '20': 'двадцать', '30': 'тридцать', '40': 'сорок', '50': 'пятьдесят',
    '60': 'шестьдесят', '70': 'семьдесят', '80': 'восемьдесят', '90': 'девяносто',
    '100': 'сто', '200': 'двести', '300': 'тристо', '400': 'четыресто',
    '500': 'пятьсот', '600': 'шестьсот', '700': 'семьсот', '800': 'восемьсот', '900': 'девятьсот'
}

dict_sig = {
    '+': 1, '-': 1, '*': 2, '/': 2, '(': 0, ')': 0
}

dict_signs = {
    'плюс': '+', 'умножить': '*', 'разделить': '/', 'минус': '-'
}

signs = ['+', '-', '*', '/', '(', ')']

def result(ch):
    ch = str(ch)
    cel, drob = ch[0:ch.find('.')], str(round(float('0.'+ch[ch.find('.') + 1:len(ch)]), 3))
    drob_itog = ''
    if cel[0] == '-':
        cel, cel_itog = cel[1:], 'минус '
    else:
        cel_itog = ''
    a = [int(cel) // 1000 * 1000, int(cel) // 100 % 10 * 100, int(cel) // 10 % 10 * 10, int(cel) % 10]
    if a[2] == 10:
        a[2], a[3] = a[2] + a[3], 0
    for elem in a:
        if elem != 0:
            cel_itog += dict_obrat_chisla[str(elem)] + ' '
    if cel_itog in ['','минус ']:
        cel_itog += 'ноль '
    drob = drob[2:len(drob)]
    if drob == '0':
        return cel_itog
    else:
        a = [int(drob) // 100 * 100, int(drob) // 10 % 10 * 10, int(drob) % 10]
        if a[2] == 10:
            a[2], a[3] = a[2] + a[3], 0
        for elem in a:
            if elem != 0:
                drob_itog += dict_obrat_chisla_drob[str(elem)] + ' '
        if len(drob) == 1:
            return cel_itog + 'и ' + drob_itog + 'десятых'
        elif len(drob) == 2:
            return cel_itog + 'и ' + drob_itog + 'сотых'
        elif len(drob) == 3:
            return cel_itog + 'и ' + drob_itog + 'тысячных'
    return 

def minus(mas):
    i = 2
    while i != len(mas):
        if (mas[i] == '-') and (mas[i - 1] in signs):
            mas[i + 1] = '-' + mas[i + 1]
            mas.pop(i)
        else:
            i += 1
    return mas
            
def chislo(mas):
    ch, dr, i = 0, 0, 0
    while mas[i] != 'и':
        ch += dict_chisla[mas[i]]
        i += 1
        if i == len(mas):
            return str(ch)
    i += 1
    while not dict_chisla[mas[i]] in [0.1, 0.01, 0.001]:
        dr += dict_chisla[mas[i]]
        i += 1
    dr = dr * dict_chisla[mas[i]]
    return str(ch + dr)

def podgot(mas):
    a, s = ['0', '+'], list()
    for i in range(0, len(mas)):
        if mas[i] in dict_chisla:
            s.append(mas[i])
        elif len(s) != 0:
            a.append(chislo(s))
            s = list()
        if mas[i] == 'открывается':
            a.append('(')
        if mas[i] == 'закрывается':
            a.append(')')
        if mas[i] in dict_signs:
            a.append(dict_signs[mas[i]])
    if len(s) != 0:
        a.append(chislo(s))
    return a
    
def delete(mas):
    a = ['на', 'скобка']
    for elem in a:
        while elem in mas:
            mas.remove(elem)
    return mas

def priority(mas):
    start = []
    end = []
    for elem in mas:
        if elem == '(':
            start = [elem] + start
        elif elem in signs:
            if start == []:
                start = [elem]
            elif elem == ')':
                while(True):
                    q = start[0]
                    start = start[1:]
                    if q == '(':
                        break
                    end += [q]
            elif prty(start[0]) < prty(elem):
                start = [elem] + start
            else:
                while(True):
                    if start == []:
                        break
                    q = start[0]
                    end += [q]
                    start = start[1:]
                    if prty(q) == prty(elem):
                        break
                start = [elem] + start
        else:
            end += [elem]
    while(start != []):
        q = start[0]
        end += [q]
        start = start[1:]
    return end

def prty(sig):
    if sig == '+' or sig == '-':
        return 1
    elif sig == '*' or sig == '/':
        return 2
    elif sig == '(' or sig == ')':
        return 0

def calc(mas):
    while len(mas) != 1:
        for i in range(len(mas)):
            if mas[i] in signs:
                mas = mas[0:i-2] + [str(eval(mas[i-2] + mas[i] + mas[i-1]))] + mas[i+1:len(mas)]
                break
    return mas
        
expression = result(float(calc(priority(minus(podgot(delete(input('Введите выражение: ').split())))))[0]))

print(expression)
```
